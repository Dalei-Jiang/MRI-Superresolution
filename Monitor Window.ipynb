```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 709
low_addr = './data/data_double_cart_64/real/low/%04d.npy'%i
res_addr = './data/data_double_cart_64/real/result/%04d.npy'%i
high_addr = './data/data_double_cart_64/real/high/%04d.npy'%i
```

```python
low = np.load(low_addr)
res = np.load(res_addr)
high = np.load(high_addr)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(low)
plt.colorbar()
plt.figure(figsize=(20,20))
plt.imshow(res)
plt.colorbar()
plt.figure(figsize=(20,20))
plt.imshow(high)
plt.colorbar()

print(np.min(low))
print(np.min(res))
print(np.min(high))
```
