```python
import argparse
import logging
import os
import numpy as np
import torch
import torch.nn.functional as F
from torchvision import transforms
from patchify import patchify, unpatchify
from math import log10, sqrt
from numpy import fft

from unet import UNet
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
```

```python
model_address = './MODEL_abs.pth'
```

```python
data_set = np.load('./T2_data/img_patchset.npy')
```

```python
i = 34
j = 38
low_patch = data_set[i,j,:,:]
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
net = UNet(n_channels=1, n_classes=1, bilinear=False)
net.load_state_dict(torch.load(model_address, map_location=device),strict=True)

pil_np = low_patch
img_ndarray = pil_np[np.newaxis, ...]
img_ndarray = img_ndarray[np.newaxis, ...]
img_ndarray = img_ndarray / 255
image = torch.as_tensor(img_ndarray.copy()).float().contiguous()
print('Image loaded...')

with torch.no_grad():
    print('Predicting the result...')
    masks_pred = net(image)
    pred = masks_pred.detach().numpy()[0, 0, :, :]
    print(np.max(pred),np.min(pred))

mod = 'rainbow'

plt.figure(figsize=(20,20))
plt.imshow(pil_np,mod,vmax=255.0)
# plt.colorbar()
plt.figure(figsize=(20,20))
plt.imshow(pred,mod,vmax=255.0)
# plt.colorbar()
```
